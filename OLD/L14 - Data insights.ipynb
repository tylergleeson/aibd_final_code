```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
```

```python
df = pd.read_csv('app_spending.csv')
```

# Box Plot

```python
# ========================================
# 2. Generate Plots for Numerical Features
# ========================================

# --------------------------
# Box Plot
# --------------------------
plt.figure(figsize=(5, 5))
sns.boxplot(data=df[["Time_Spent"]], showmeans=True,
            meanprops={'marker': 'o', 'markerfacecolor': 'white', 'markeredgecolor': 'black', 'markersize': '8'})
plt.title("Box Plots: Time Spent")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np



# Set custom colors for Y = 0 group and Y = 1 group
palette = {0: 'skyblue', 1: 'salmon'}

# Seaborn boxplot
sns.boxplot(data=df, x='Y', y='Time_Spent',hue='Y', palette=palette, showmeans=True)

# use pre-defined palette
# sns.boxplot(data=df, x='Y', y='Time_Spent',hue='Y', palette="muted", showmeans=True)
# Categorical palettes:
# "deep" (default)
# "muted"
# "bright"
# "pastel"
# "dark"

plt.title("Time Spent by Y Value")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np



# Set custom colors for Y = 0 group and Y = 1 group
palette = {0: 'skyblue', 1: 'salmon'}

# Seaborn boxplot
sns.boxplot(data=df, x='Age_Group', y='Subscribed_Creators_Count',hue='Age_Group', palette="muted", showmeans=True)

# use pre-defined palette
# sns.boxplot(data=df, x='Y', y='Time_Spent',hue='Y', palette="muted", showmeans=True)
# Categorical palettes:
# "deep" (default)
# "muted"
# "bright"
# "pastel"
# "dark"

plt.title("Subscribed Creator Count by Y Value")
plt.legend()
plt.show()
```

# Histogram

```python


sns.histplot(data=df, x='Subscribed_Creators_Count', hue='Age_Group', kde=True, element='step', stat='count', common_norm=False, alpha=0.6)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: Compare Monthly Short Video Views by Age Group


fig, ax = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(data=df, x='Monthly_Short_Video_Views', hue='Age_Group', kde=False, element='step', stat='count', common_norm=False, alpha=0.6, ax=ax[0])
sns.histplot(data=df, x='Monthly_Long_Video_Views', hue='Age_Group', kde=False, element='step', stat='count', common_norm=False, alpha=0.6, ax=ax[1])
ax[0].set_title("The numbers of short videos viewed per month")
ax[1].set_title("The numbers of long videos viewed per month")
plt.ylabel('Count')
plt.show()
```

# Bar Plot

```python

# ----------------------------
# Categorical Attributes (Single)
# ----------------------------
# Bar Plot
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
# manually define the color
sns.countplot(data=df, x="Age_Group", ax=ax[0], palette={"Young": "skyblue", "Senior": "salmon"})
# use pre-defined palattes
sns.countplot(data=df, x="Membership_Type", ax=ax[1], palette = "muted")


# "colorblind"
ax[0].set_title("Age Group Distribution")
ax[1].set_title("Membership Type Distribution")
plt.show()
```

# Mosaic Plot

```python
plt.figure(figsize=(8, 4))
mosaic(df, ["Age_Group", "Membership_Type"], title="Mosaic Plot for Age Group and Membership Type")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
mosaic(df, ["Age_Group","Y"], title="Mosaic Plot for Age_Group and Y")
plt.show()
```

```python
# ----------------------------
# Categorical Attributes (Pairs)
# ----------------------------


# ================================
# 3. Correlation ≠ Causation Analysis
# ================================
# Key visualization for discussion
plt.figure(figsize=(6, 5))
sns.scatterplot(data=df, x="Time_Spent", y="Monthly_Short_Video_Views")
plt.title("Time Spent vs Monthly_Short_Video_Views")
plt.show()
```

```python

plt.figure(figsize=(6, 5))
sns.scatterplot(data=df, x="Time_Spent", y="Monthly_Short_Video_Views", hue="Age_Group")
plt.title("Time Spent vs Monthly_Short_Video_Views")
plt.show()
```
